```python
import matplotlib.pyplot as plt
import matplotlib as mpl

fig, ax = plt.subplots() # 创建图实例
x = [-4,-3,-2,-1,0,1] # 创建x的取值范围


y1 = [0.48571429, 0.48571429, 0.80952381, 0.95238095, 0.98095238, 0.94285714]
ax.plot(x, y1, label='linear')
 
y2 = [0.43809524, 0.68571429, 0.9047619, 0.98095238, 0.96190476, 0.92380952]
ax.plot(x, y2, label='poly')

y3 = [0.48571429, 0.48571429, 0.48571429,0.79047619,  0.96190476,  0.97142857]
ax.plot(x, y3, label='rbf') 

y4 = [0.23809524, 0.23809524, 0.23809524, 0.23809524, 0.23809524,  0.1047619]
ax.plot(x, y4, label='sigmoid') 

ax.set_xlabel('Parameter log10C') #设置x轴名称 x label
ax.set_ylabel('Predict Acc') #设置y轴名称 y label
ax.set_title('Fine-tune C and kernel') #设置图名为Simple Plot
ax.legend() #自动检测要在图例中显示的元素，并且显示

plt.show() #图形可视化
```
