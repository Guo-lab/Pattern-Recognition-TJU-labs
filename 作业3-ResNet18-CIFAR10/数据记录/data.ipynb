```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.loadtxt('./3-2/5.txt')
y1 = np.loadtxt('./3-2/7.txt')
y2 = np.loadtxt('./4-2/7.txt')
y3 = np.loadtxt('./5/7.txt')
y4 = np.loadtxt('./2/7.txt')
y5 = np.loadtxt('./1/7.txt')

fig = plt.figure()

plt.plot(x, y1, label="Normalize+RandomGrayscale+ColorJitter")
plt.plot(x, y2, label="Normalize+Crop+Flip+Rotation")
plt.plot(x, y3, label="Normalize+all")
plt.plot(x, y4, label="Normalize")
plt.plot(x, y5, label="Raw Data in Tensor", linewidth=3)

plt.title("Compare Data Augmentation")
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Test Acc')

plt.show()
```

```python
similarity = [73.60,		75.3,74.03,	73.00,	67.21]  # similarity of action
divergence = [1898.681115,		1371.89886069297, 1150.083187,	1014.135447,	983.8010662]  # js diversity
labels = ['8',	'16',	'32',	'128',	'512']

#plt.rcParams['axes.labelsize'] = 16  # xy轴label的size
#plt.rcParams['xtick.labelsize'] = 12  # x轴ticks的size
#plt.rcParams['ytick.labelsize'] = 14  # y轴ticks的size
# plt.rcParams['legend.fontsize'] = 12  # 图例的size

# 设置柱形的间隔
width = 0.3  # 柱形的宽度
x1_list = []
x2_list = []
for i in range(len(similarity)):
    x1_list.append(i)
    x2_list.append(i + width)

# 创建图层
fig, ax1 = plt.subplots()

# 设置左侧Y轴对应的figure
ax1.set_ylabel('Test accuracy (%) epoch = 10')
ax1.set_ylim(65, 76)
ax1.bar(x1_list, similarity, width=width, alpha=0.7, align='edge')

ax1.set_xticklabels(ax1.get_xticklabels())  # 设置共用的x轴

# 设置右侧Y轴对应的figure
ax2 = ax1.twinx()
ax2.set_ylabel('Running time (s)')
ax2.set_ylim(0, 2000)
ax2.bar(x2_list, divergence, width=width, color='indianred', alpha=0.7, align='edge', tick_label=labels)

plt.tight_layout()
plt.savefig("Comparison - Batch Size.png")
plt.show()
```
